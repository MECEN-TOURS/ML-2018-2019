## Pour rajouter une ligne

```python
df = pd.DataFrame(np.random.rand(30, 5), columns=list("abcde"))
df
```

```python
def moyenne(x):
    return np.mean(x)
```

```python
df.loc["moyenne"] = df.apply(moyenne)
```

```python
df
```

**Pour se débarrasser d'une ligne ou d'une colonne**

```python
df = df.drop(axis="index", labels="moyenne")
df
```

## Remarque  
La majorité des opérations renvoient un nouvel objet plutôt que de muter l'objet sous jacent.  
L'intérêt est de pouvoir chainer les méthodes.  

```python
df = df.apply(lambda x:np.ceil(6*x)).astype(np.int)
df.head()
```
