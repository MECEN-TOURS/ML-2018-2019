```python
s_l = [x+y for x,y in zip(x_l, y_l)]
print(type(s_l), s_l)

s_t = x_t+y_t
print(type(s_t), s_t)
```

```python
p_l = [x*y for x,y in zip(x_l, y_l)]
print(type(p_l), p_l)

p_t = x_t*y_t
print(type(p_t), p_t)
```

```python
c_l = [x**2 for x in x_l]
print(type(c_l), c_l)

u_t = x_t**2
print(type(u_t), u_t)
```

```python
b_l = [x<y for x,y in zip(x_l, y_l)]
print(b_l)
b_t = x_t < y_t
print(b_t)
```

# Un exemple plus sophistiqué : les ensembles de Julia
## Exécuter le code suivant.

**Attention** en python le nombre $i$ s'obtient en fait par `1j` ou aussi `complex(0,1)`.

```python
def julia(z):
    return z**2+0.285+0.01j

x_m, x_M, y_m, y_M = -1, 1, -1, 1
largeur, hauteur = 400, 400
x = np.linspace(x_m, x_M, largeur)
y = np.linspace(y_m, y_M, hauteur)

# deux lignes plus subtiles, à analyser en dernier
z = x[np.newaxis, :] +1j*y[:, np.newaxis]
iterations = np.zeros_like(z, dtype=np.int)

lim = 10
iterations_max = 100
for i in range(iterations_max):
    bornes = np.abs(z)<lim
    iterations[bornes] += 1
    z[bornes] = julia(z[bornes])

plt.imshow(iterations, 
           extent=(x_m, x_M, y_m, y_M),
           cmap="viridis"
           );
```

## Version pile ou face

```python
def une_partie(fortune_ini):
    duree, fortune = 0, fortune_ini
    while 0<fortune<100:
        duree += 1
        p = np.random.random()
        if p<0.5:
            fortune = fortune-1
        else:
            fortune = fortune+1
            
    return fortune, duree
```

```python
une_partie(50)
```

```python
def estimation_loi(fortune_ini, nb_de_parties):
    fortunes, durees = list(), list()
    for i in range(nb_de_parties):
        fortune, duree = une_partie(fortune_ini)
        fortunes.append(fortune)
        durees.append(duree)
    return fortunes, durees
```

Pour calculer la probabilité de gagner en fonction de la fortune initiale on calcule la moyenne empirique.

```python
def chance_de_gagner(nb_de_parties=1000):
    mises = np.arange(0, 101)
    moyenne_gains = list()
    moyenne_durees = list()
    for mise in mises:
        gains, durees = estimation_loi(mise, nb_de_parties)
        moyenne_gains.append(sum(gains)/nb_de_parties)
        moyenne_durees.append(sum(durees)/nb_de_parties)
    
    fig, ax = plt.subplots(1, 2)
    ax_g, ax_d = ax
    ax_g.plot(mises, moyenne_gains)
    ax_g.set_title("Moyenne des gains en fonction de la mise initiale")
    ax_d.plot(mises, moyenne_durees)
    ax_d.set_title("Moyenne des durées de jeux en fonction de la mise initiale")
        
        
```

## Roulette

Dans le jeu de la roulette on à 37 cases : 18 rouges, 18 noires et 1 verte. Le joueur peut parier sur une couleur pour le prochain lancer et à alors une probabilité $\frac{18}{37}$ de tomber juste s'il a choisit rouge ou noir. Cela correspond à un pile ou face pipé on peut donc réutiliser le code précédent. (On suppose qu'on ne joue pas tout son argent en une fois)

```python
def roulette_une_partie(fortune_ini):
    duree, fortune = 0, fortune_ini
    while 0<fortune<100:
        duree += 1
        p = np.random.random()
        if p<19/37:
            fortune = fortune-1
        else:
            fortune = fortune+1
            
    return fortune, duree
```

```python
def roulette_estimation_loi(fortune_ini, nb_de_parties):
    fortunes, durees = list(), list()
    for i in range(nb_de_parties):
        fortune, duree = roulette_une_partie(fortune_ini)
        fortunes.append(fortune)
        durees.append(duree)
    return fortunes, durees
```

```python
def roulette_affichage(fortune_ini, nb_de_parties):
    fortunes, durees = roulette_estimation_loi(fortune_ini, nb_de_parties)
    fig, ax = plt.subplots(1,2)
    ax_gauche, ax_droite = ax
    ax_gauche.hist(fortunes)
    ax_gauche.set_title("Répartition des fortunes terminales")
    ax_droite.hist(durees)
    ax_droite.set_title("Répartition du nombre de pile ou face par partie")
```

```python
roulette_affichage(50, 1000)
```

On voit des distributions assez biaisées. On va là encore calculer les moyennes empiriques en fonction de la mise initiale.

```python
def roulette_chance_de_gagner(nb_de_parties=100):
    mises = np.arange(0, 101)
    moyenne_gains = list()
    moyenne_durees = list()
    for mise in mises:
        gains, durees = roulette_estimation_loi(mise, nb_de_parties)
        moyenne_gains.append(sum(gains)/nb_de_parties)
        moyenne_durees.append(sum(durees)/nb_de_parties)
    
    fig, ax = plt.subplots(1, 2)
    ax_g, ax_d = ax
    ax_g.plot(mises, moyenne_gains)
    ax_g.set_title("Moyenne des gains en fonction de la mise initiale")
    ax_d.plot(mises, moyenne_durees)
    ax_d.set_title("Moyenne des durées de jeux en fonction de la mise initiale")
```

```python
%%time
roulette_chance_de_gagner()
```

**Conclusion** on voit que le fait d'utiliser une pièce biaisée même de peu change radicalement la fonction mise initiale -> probablité de gagner.

## Accélération du code

A cause des fluctuations des durées de partie, la vectorisation du code n'est pas évidente et surtout n'engendrerait pas de gain énorme sur la vitesse d'exécution on va utiliser le module `cython` qui permet en rajoutant des annotations de types d'accélérer fortement l'exécution de code.

```python
%load_ext cython
```

```python
%%cython -a

import numpy 
cimport numpy

import cython

from libc.stdlib cimport rand, RAND_MAX

@cython.cdivision(True)
cdef double random():
    return rand()/float(RAND_MAX)


cdef (double, double) une_partie_cy(double fortune_ini):
    cdef double duree=0, fortune=fortune_ini
    while 0<fortune<100:
        duree += 1
        p = random()
        if p<0.5:
            fortune = fortune-1
        else:
            fortune = fortune+1
            
    return fortune, duree

@cython.cdivision(True)
cdef (double, double) moyenne_empirique_cy(double fortune_ini, int nb_de_parties):
    cdef double moyenne_f=0, moyenne_d=0
    cdef double fortune, duree
    cdef cython.int i
    for i in range(nb_de_parties):
        fortune, duree = une_partie_cy(fortune_ini)
        moyenne_f = moyenne_f + fortune
        moyenne_d = moyenne_d + duree
    return moyenne_f/float(nb_de_parties), moyenne_d/float(nb_de_parties)

@cython.boundscheck(False)
def chance_de_gagner(nb_de_parties=100):
    cdef double [:] mises = numpy.arange(101, dtype=numpy.float)
    cdef double [:] moyennes_gains = numpy.zeros(shape=(101))
    cdef double [:] moyennes_durees = numpy.zeros(shape=(101))
    cdef Py_ssize_t i
    for i in range(101):
        moyenne_gains, moyenne_durees = moyenne_empirique_cy(mises[i], nb_de_parties)
        moyennes_gains[i] = moyenne_gains
        moyennes_durees[i] = moyenne_durees
    return moyennes_gains, moyennes_durees
    
```

```python
%%time
moyennes_gains, moyennes_durees = chance_de_gagner()
```

```python
mises = np.arange(101)
fig, ax = plt.subplots(1, 2)
ax_g, ax_d = ax
ax_g.plot(mises, moyennes_gains)
ax_g.set_title("Moyenne des gains en fonction de la mise initiale")
ax_d.plot(mises, moyennes_durees)
ax_d.set_title("Moyenne des durées de jeux en fonction de la mise initiale")
```
